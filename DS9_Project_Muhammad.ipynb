```python
# Standard Library
import pandas as pd
import numpy as np

# Matplotlib
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 8)
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
plt.style.use('ggplot')

# Seaborn
import seaborn as sns
```

```python
df = pd.read_csv('C://Users//HP//Python//Dataset//COVID-19 Case Data//National Data//Indonesia_coronavirus_daily_data.csv')
df
```

```python
df.Province.unique()
```

```python
DKI_Jakarta = df[df['Province'] == 'DKI JAKARTA']
DKI_Jakarta
```

```python
Papua = df[df['Province'] == 'PAPUA']
Papua
```

fokus utamanya adalah visualisasi time series

```python
fig, ax = plt.subplots(ncols = 2,
                       figsize=(24,16))

sns.lineplot(x = "Date", 
             y = "Daily_Case",
             data = DKI_Jakarta, ax=ax[0], label = 'Daily_Case')
sns.lineplot(x = "Date", 
             y = "Daily_Death",
             data = DKI_Jakarta, ax=ax[0], label = 'Daily_Death')
sns.lineplot(x = "Date", 
             y = "Daily_Recovered",
             data = DKI_Jakarta, ax=ax[0], label = 'Daily_Recovered')

loc = MultipleLocator(base=14)
ax[0].xaxis.set_major_locator(loc)
ax[0].set(xlabel = "Date", ylabel = "Cases")
ax[0].set_title('DKI Jakarta', size=20)

sns.lineplot(x = "Date", 
             y = "Daily_Case",
             data = Papua, ax=ax[1], label = 'Daily_Case')
sns.lineplot(x = "Date", 
             y = "Daily_Death",
             data = Papua, ax=ax[1], label = 'Daily_Death')
sns.lineplot(x = "Date", 
             y = "Daily_Recovered",
             data = Papua, ax=ax[1], label = 'Daily_Recovered')

loc = MultipleLocator(base=14)
ax[1].xaxis.set_major_locator(loc)
ax[1].set(xlabel = "Date", ylabel = "Cases")
ax[1].set_title('Papua', size=20)

ax[0].tick_params(axis='x', labelrotation=45)
ax[1].tick_params(axis='x', labelrotation=45)
```

```python
fig, axc = plt.subplots(ncols = 2,
                       figsize=(24,16))

sns.lineplot(x = "Date", 
             y = "Cumulative_Case",
             data = DKI_Jakarta, ax=axc[0], label = 'Cumulative_Case')
sns.lineplot(x = "Date", 
             y = "Cumulative_Death",
             data = DKI_Jakarta, ax=axc[0], label = 'Cumulative_Death')
sns.lineplot(x = "Date", 
             y = "Cumulative_Recovered",
             data = DKI_Jakarta, ax=axc[0], label = 'Cumulative_Recovered')
sns.lineplot(x = "Date", 
             y = "Cumulative_Active_Case",
             data = DKI_Jakarta, ax=axc[0], label = 'Cumulative_Active_Case')


loc = MultipleLocator(base=28)
axc[0].xaxis.set_major_locator(loc)
axc[0].set(xlabel = "Date", ylabel = "Cases")
axc[0].set_title('DKI Jakarta', size=20)

sns.lineplot(x = "Date", 
             y = "Cumulative_Case",
             data = Papua, ax=axc[1], label = 'Cumulative_Case')
sns.lineplot(x = "Date", 
             y = "Cumulative_Death",
             data = Papua, ax=axc[1], label = 'Cumulative_Death')
sns.lineplot(x = "Date", 
             y = "Cumulative_Recovered",
             data = Papua, ax=axc[1], label = 'Cumulative_Recovered')
sns.lineplot(x = "Date", 
             y = "Cumulative_Active_Case",
             data = Papua, ax=axc[1], label = 'Cumulative_Active_Case')


loc = MultipleLocator(base=28)
axc[1].xaxis.set_major_locator(loc)
axc[1].set(xlabel = "Date", ylabel = "Cases")
axc[1].set_title('Papua', size=20)

axc[0].tick_params(axis='x', labelrotation=45)
axc[1].tick_params(axis='x', labelrotation=45)
```
